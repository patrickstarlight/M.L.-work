```python
  import sklearn
from pandas.core.dtypes.common import validate_all_hashable
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import metrics 
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression as LogR
from sklearn.datasets import make_blobs as sam
from sklearn.preprocessing import MinMaxScaler as MMS
from sklearn.preprocessing import LabelEncoder 
from sklearn.compose import ColumnTransformer as CTF
import statsmodels.api as sts
input_data = pd.read_csv("/content/drive/MyDrive/466920-2021.csv",delimiter=",",header=0)     
data=input_data.iloc[1:,[7,8,10,12,19,20,27]].values
label=input_data.iloc[1:,18].values
print(data)
print(label)
Train_D,Test_D,Train_L,Test_L = train_test_split(data,label,test_size=0.1,random_state=0)
#print(Train_D[0:30,])
#print(Train_L[0:30,])
from sklearn.preprocessing import StandardScaler
fS = StandardScaler()
print("---Training---")
Train_D=fS.fit_transform(Train_D)
Test_D=fS.fit_transform(Train_D)

print("")
print("---Testing---")

print(Test_D[0:30,])
cls = LogR(random_state=0)
cls.fit(Train_D,Train_L)
pred = cls.predict(Test_D)

print(pred[:30,])
class_labels = LabelEncoder()
Test_L = class_labels.fit_transform(Test_L)
y_perd_proba = cls.predict_proba(Test_D)[::,0]
fpr,tpr,_=metrics.roc_curve(Test_L,y_perd_proba)
auc = metrics.roc_auc_score(Test_L,y_perd_proba)
auc = str(round(auc,2))

plt.plot(fpr,tpr,label="auc="+str(auc))
plt.plot([0,1],[0,1],color="red",linestyle="-")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```
